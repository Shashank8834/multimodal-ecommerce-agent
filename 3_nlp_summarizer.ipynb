```python
from transformers import pipeline
import pandas as pd
```

```python
products = [
    {
        "title": "Faux Leather Jacket",
        "price": 89.99,
        "description": "This faux leather jacket is perfect for casual wear. Stylish and comfortable."
    },
    {
        "title": "Cotton T-Shirt",
        "price": 19.99,
        "description": "Soft cotton t-shirt with a relaxed fit, great for everyday use."
    },
    {
        "title": "Sports Backpack",
        "price": 49.99,
        "description": "Durable sports backpack with multiple compartments and padded straps."
    }
]

df = pd.DataFrame(products)
```

```python
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
```

```python
def summarize_product(title, description):
    text = f"Product: {title}. Details: {description}"
    # Hugging Face summarizers usually expect longer text, so concatenate
    summary = summarizer(text, max_length=60, min_length=20, do_sample=False)[0]['summary_text']
    return summary
```

```python
df['summary'] = df.apply(lambda row: summarize_product(row['title'], row['description']), axis=1)
```

```python
for idx, row in df.iterrows():
    print(f"{idx+1}. {row['title']} (${row['price']})")
    print(f"Summary: {row['summary']}\n")
```
